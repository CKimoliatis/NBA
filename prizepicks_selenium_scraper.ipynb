```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()

driver.get("https://parlayplay.io/")
print("🟡 Waiting for page to fully load...")
time.sleep(10)

# ✅ Step 1: Click the Rebounds tab
try:
    rebounds_tab = driver.find_element(By.XPATH, "//button[.//span[text()='Points']]")
    rebounds_tab.click()
    print("✅ Clicked on the Points tab.")
    time.sleep(5)  # Give time for Rebounds data to load
except Exception as e:
    print(f"❌ Couldn't click Points tab: {e}")
    driver.quit()
    exit()


players = driver.find_elements(By.XPATH, "//*[starts-with(@id, 'player-')]")
print(f"🟡 Found {len(players)} player elements")

data = []

for idx, player in enumerate(players):
    try:
        # ✅ Name
        name = player.find_element(By.CSS_SELECTOR, "div.w-full.mx-auto.mt-2.text-center.text-black div.px-3.font-display").text.strip()

        if name == "" or "Boost" in name:
            print(f"⏩ Skipping invalid player: {name}")
            continue

        # ✅ Stat Value
        stat_value = player.find_element(By.CSS_SELECTOR, "[data-testid='stat-value']").text.strip()

        # ✅ Stat Type
        stat_type = player.find_element(By.CLASS_NAME, "text-xs").text.strip()

        # ✅ Matchup and Game Time
        matchup_block = player.find_element(By.CSS_SELECTOR, "div.flex-row.items-center.text-xs")
        spans = matchup_block.find_elements(By.TAG_NAME, "span")
        matchup = spans[0].text.strip()
        game_time = spans[1].text.strip()

        # ✅ Odds
        odds_buttons = player.find_elements(By.CSS_SELECTOR, "[data-testid='grid-button']")
        odds = [btn.text.strip() for btn in odds_buttons]

        data.append({
            "Player": name,
            "Stat Type": stat_type,
            "Line": stat_value,
            "Matchup": matchup,
            "Game Time": game_time,
            "Odds": ", ".join(odds)
        })

        print(f"{idx+1}. {name} | {stat_type} | {stat_value} | {matchup} | {game_time} | Odds: {', '.join(odds)}")

    except Exception as e:
        print(f"❌ Error on player {idx+1}: {e}")

# ✅ Save to CSV
df = pd.DataFrame(data)
df.to_csv("parlayplay_points.csv", index=False)
df.head()

print(f"\n✅ Scraped {len(df)} valid players and saved to parlayplay_props.csv.")
```

```python
import pandas as pd
import os
from datetime import datetime

# ✅ Load the CSVs
points_df = pd.read_csv("parlayplay_points.csv")
rebounds_df = pd.read_csv("parlayplay_rebounds.csv")
assists_df = pd.read_csv("parlayplay_assists.csv")

# ✅ Today's date (change if scraping for future dates)
today = datetime.today().strftime('%Y-%m-%d')

# ✅ Collect all unique players
players = set(points_df['Player']).union(rebounds_df['Player']).union(assists_df['Player'])

# ✅ Make sure output folder exists
os.makedirs("players", exist_ok=True)

for player in players:
    # Get player data from each CSV
    player_points = points_df[points_df['Player'] == player]
    player_rebounds = rebounds_df[rebounds_df['Player'] == player]
    player_assists = assists_df[assists_df['Player'] == player]
    
    # Get matchup and opponent
    matchup = ""
    if not player_points.empty:
        matchup = player_points.iloc[0]["Matchup"]
    elif not player_rebounds.empty:
        matchup = player_rebounds.iloc[0]["Matchup"]
    elif not player_assists.empty:
        matchup = player_assists.iloc[0]["Matchup"]
    
    opponent = matchup.split("@")[1] if "@" in matchup else ""

    # Get odds (grab from first available)
    odds = ""
    if not player_points.empty:
        odds = player_points.iloc[0]["Odds"]
    elif not player_rebounds.empty:
        odds = player_rebounds.iloc[0]["Odds"]
    elif not player_assists.empty:
        odds = player_assists.iloc[0]["Odds"]

    # Build the row of data
    new_row = {
        "Date": today,
        "Points": player_points["Line"].values[0] if not player_points.empty else "",
        "Rebounds": player_rebounds["Line"].values[0] if not player_rebounds.empty else "",
        "Assists": player_assists["Line"].values[0] if not player_assists.empty else "",
        "Opponent": opponent,
        "Odds": odds
    }

    filename = f"players/{player.replace(' ', '_')}.csv"

    # ✅ If file exists, append if today's data isn't there yet
    if os.path.exists(filename):
        player_df = pd.read_csv(filename)
        if today not in player_df["Date"].values:
            player_df = pd.concat([player_df, pd.DataFrame([new_row])], ignore_index=True)
            player_df.to_csv(filename, index=False)
            print(f"✅ Updated {filename} with new data.")
        else:
            print(f"🔁 {filename} already has today's data.")
    else:
        # ✅ Create new player CSV
        pd.DataFrame([new_row]).to_csv(filename, index=False)
        print(f"✅ Created new file: {filename}")

print("\n🎉 All player CSVs updated with Odds!")
```
